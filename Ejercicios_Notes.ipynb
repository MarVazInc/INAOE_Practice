# 🟢 Ejercicio 1 — Búsqueda Lineal (Calentamiento)

### 📌 Instrucción

Implementa una función que:

- reciba una lista `arr`
- reciba un valor `x`
- regrese `True` si existe, `False` si no

---

### 🧠 Qué debes practicar

- ciclos (`for`)
- comparación
- lógica básica

---

### ⚠️ Restricción

❌ No usar `in`  
❌ No copiar código  

---

````python
def busqueda_lineal(arr, x):
    for i in arr:
        if i == x:
            return True
    return False

print(busqueda_lineal([1, 2, 3, 4, 5],5))
print(busqueda_lineal([1,2,4,6,7],5))
````

### 📌 Instrucción

Encuentra el valor máximo de una lista.

---

### 🧠 Qué debes practicar

- recorrido
- comparación acumulativa

---

### 💡 Hint

Mantén una variable `max_actual`

---

````python
def maximo_de_arreglo(arr):
    max_actual = arr[0]

    for i in range(len(arr)):
         if arr[i] > max_actual:
             max_actual = arr[i]
    return max_actual

print(maximo_de_arreglo([1,2,3,4,5,6,7]))
````

````python
def maximo(arr):
    max_actual = arr[0]

    for elemento in arr:
        if elemento > max_actual:
            max_actual = elemento
    return max_actual


print(maximo([3,6,1,2,8, 13, 2]))
````


# 🔴 Ejercicio 3 — Recursión: Factorial

### 📌 Instrucción

Implementa:

$$
n! = n \cdot (n-1)!
$$

con:

$$
0! = 1
$$

---

### 🧠 Qué debes practicar

- llamada a sí mismo
- caso base vs caso recursivo

---

### ⚠️ Obligatorio

Debes poder explicar:

👉 “¿cuándo se detiene?”

---

````python
def factorial(n):    
    if n == 0:
        return 1 
    return n * factorial(n-1)

print(factorial(7))
````

La recursión utiliza la pila de llamadas del sistema.
Cada llamada agrega un frame.
Sin caso base, se produce un desbordamiento de pila.



# 🔴 Ejercicio 4 — Recursión: Fibonacci

### 📌 Instrucción

Implementa Fibonacci recursivo:

$$
F(n) = F(n-1) + F(n-2)
$$

---

### 🧠 Qué debes notar

👉 Es lento → aquí conectas con $O(2^n)$

---

### 💡 Reflexión

¿Por qué es ineficiente?

---

````python
import time

def fibonacci_recursivo(n):

    inicio = time.time()

    if n == 0:
        return 0
    if n == 1:
        return 1
    
    return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)


print(fibonacci_recursivo(0))
print(fibonacci_recursivo(1))
print(fibonacci_recursivo(2))
print(fibonacci_recursivo(3))
print(fibonacci_recursivo(4))
print(fibonacci_recursivo(5))
print(fibonacci_recursivo(6))
print(fibonacci_recursivo(7))
print(fibonacci_recursivo(8))
print(fibonacci_recursivo(35))
print(fibonacci_recursivo(40))
````

F(5)
├── F(4)
│   ├── F(3)
│   │   ├── F(2)
│   │   └── F(1)
│   └── F(2)
└── F(3)
    ├── F(2)
    └── F(1)

¿Cómo evitamos que F(3) se calcule varias veces?
Este algoritmo es muy ineficiente.

Memorizando los resultados calculados.

Si ya se calculo algo, se guarda y se utiliza luego.

````python
def fibonacci_memo(n, memo=None):  # memo = {} es peligroso, usar None

    if memo is None:
        memo = {}
        
    # Si ya lo calculé, lo regreso revisando la lista ordenada.
    if n in memo:
        return memo[n]
    
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    #calculo recursivo con resultado guardado en la lista ordenada.
    #se van acumulando en la lista los calculos ya realizados.
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)

    return memo[n]


print(fibonacci_memo(0))
print(fibonacci_memo(1))
print(fibonacci_memo(2))
print(fibonacci_memo(3))
print(fibonacci_memo(4))
print(fibonacci_memo(5))
print(fibonacci_memo(35))
print(fibonacci_memo(100))
````

F(3) → se calcula una vez ✔

→ se guarda en memo

→ cuando se necesita otra vez → se reutiliza

pasa de O(2^n) a O(n)

Recomputación a Dynamic Programming.

🧠 Forma mental correcta

problema grande
+ → subproblemas
+ → guardo resultados
+ → evito repetir

````python
def fibonacci_iterativo(n):
    a, b = 0, 1     #se inicializan las variables.

    for _ in range(n):    # _ significa no me importa la variable.
        a, b = b, a + b     # se itera la suma consecutiva. a= F(k) b = F(k+1)

    return a

print(fibonacci_iterativo(0))
print(fibonacci_iterativo(1))
print(fibonacci_iterativo(2))
print(fibonacci_iterativo(3))
print(fibonacci_iterativo(4))
print(fibonacci_iterativo(5))
print(fibonacci_iterativo(35))
print(fibonacci_iterativo(100))
````

🧠 Comparación final
Método	Idea	Nivel
Recursivo	definición matemática	🟢
Memoización	optimización	🟡
Iterativo	eficiencia total	🔥

✔ Iterativo
usa memoria constante → O(1)
❌ Recursivo
usa stack → O(n)
🟡 Memoización
usa diccionario → O(n)




# 🟣 Ejercicio 5 — Búsqueda Binaria (Clave)

### 📌 Instrucción

Dado un arreglo ORDENADO:

```text
[1, 2, 3, 4, 5, 6, 7]

Implementa búsqueda binaria.

---

🧠 Variables clave
left
right
mid

---

🧪 Prueba
buscar 4 → True
buscar 10 → False

````python
def busqueda_binaria(arr, x):

    left = 0                # porque cero funciona y no arr[0]? left es índice no valor.
    right = len(arr) - 1    # porque no puede ser así sin el -1? índices son 0 al 6 no 7.

    while left <= right:
        mid = (left + right) // 2  # como producto es primero requiere paréntesis

        if arr[mid] == x:
            return True
        elif arr[mid] < x:
            left    = mid + 1
        else:
            right   = mid - 1
        return False            # por que el return no es bajo el while? 
                                #return bajo while el loop solo se ejecuta una vez
# 
print(busqueda_binaria([1,2,3,4,5,6,7],4))
print(busqueda_binaria([1,2,3,4,5,6,7],10))
````

n → n/2 → n/4 → n/8 ...

Este método es O(log n)

Duda	Respuesta

left = 0	índice

right = len(arr)-1	evitar out of bounds

mid	necesita paréntesis

return False	fuera del while

# 🧠 Ejercicio 6 — Búsqueda Binaria (Reflexión corregida)

---

## 1. ¿Por qué búsqueda binaria es más rápida que lineal?

**Respuesta refinada:**

Porque en cada iteración reduce el espacio de búsqueda a la mitad.

En lugar de revisar todos los elementos (O(n)), descarta la mitad en cada paso, logrando complejidad:

👉 O(log n)

---

## 2. ¿Cuándo NO puedes usar búsqueda binaria?

- Cuando el arreglo NO está ordenado ❌
- Cuando los datos cambian constantemente (no hay orden estable)
- Cuando necesitas encontrar múltiples ocurrencias sin adaptación del algoritmo

---

## 3. ¿Qué condición debe cumplir el arreglo?

👉 Debe estar **ordenado** (ascendente o descendente)

Esto es necesario para poder decidir:

- ir a la izquierda
- ir a la derecha

---

## 4. ¿Qué pasa si el arreglo NO está ordenado?

La búsqueda binaria:

- puede fallar
- puede descartar el elemento correcto
- produce resultados incorrectos

👉 No es confiable

---

## 🧠 Insight clave

La búsqueda binaria no “busca más rápido” por sí misma.

👉 Es rápida porque usa **estructura (orden)**

---

## 🔥 Comparación mental

| Método | Estrategia | Complejidad |
|-------|----------|------------|
| Búsqueda lineal | revisar uno por uno | O(n) |
| Búsqueda binaria | dividir el problema | O(log n) |

---

## 🎯 Idea central

👉 "Orden + División = Eficiencia"

---

## 🧠 Aprendizaje

- Arrays → acceso directo
- Recursión → dividir problema
- Fibonacci → costo de repetir
- Binaria → evitar trabajo innecesario

💥 Todo está conectado

## 🔵 Ejercicio 11 — Inserción al final

---

🧠 Objetivo

Aprender a:

recorrer + modificar el último nodo
💡 Idea clave
No tienes acceso directo al final
→ tienes que encontrarlo caminando
🔥 Visual

Antes:

5 → 10 → 20 → 30 → NULL

Insertar 99:

5 → 10 → 20 → 30 → 99 → NULL

Paso 1 — Crear nodo nuevo

Debajo de lo que ya tienes:

Node* nuevo_final = new Node();
nuevo_final->data = 99;
nuevo_final->next = NULL;
✍️ Paso 2 — Encontrar el último nodo
Node* temp = head;
🧠 Este puntero es CLAVE

👉 temp va a caminar hasta el final
👉 sin perder head

✍️ Paso 3 — Recorrer hasta el final
while (temp->next != NULL) {
    temp = temp->next;
}
🧠 ¿Por qué temp->next != NULL?

Porque:

el último nodo es el que apunta a NULL
✍️ Paso 4 — Insertar
temp->next = nuevo_final;
🚀 Paso 5 — Imprimir otra vez

Reutiliza tu while:

Node* current = head;

while (current != NULL) {
    cout << current->data << endl;
    current = current->next;
}

⚠️ Error típico (importante)

❌ hacer esto:

while (temp != NULL)

👉 te pasas del último nodo
👉 pierdes la referencia

🧠 Insight fuerte
Insertar al inicio = cambiar head
Insertar al final  = encontrar último nodo
💥 Lo que acabas de aprender
recorrido con propósito ✔
modificación estructural ✔
navegación segura ✔

🧠 Lo más importante que ya dominaste aquí
1. new → memoria dinámica (heap)
2. Node* → direcciones, no valores
3. head → punto de entrada, no la lista
4. Inserción inicio → cambia head
5. Inserción final → requiere recorrido
6. while → solo para buscar, NO modificar
7. NULL → fin de estructura


🔥 Insight nivel entrevista (esto ya es serio)
Array:
- memoria contigua
- acceso O(1)

Linked List:
- memoria dispersa
- acceso O(n)
- inserción flexible

## 🔵 Ejercicio 12 — Búsqueda en lista

---
List
🧠 Objetivo

Implementar búsqueda en una estructura dinámica.

💡 Idea clave
No hay índices → tienes que recorrer

🔥 Diferencia clave
Array → acceso directo
Lista → acceso secuencial

🎯 Tarea

Buscar si existe un valor en la lista.

Ejemplo:

Buscar 20 → TRUE
Buscar 100 → FALSE

🧠 Lógica esperada
+ Empiezas en head
+ Comparas data
+ Avanzas con next

Si encuentras → TRUE

Si llegas a NULL → FALSE

⚠️ Restricción mental
+ NO puedes saltar nodos
+ NO puedes usar índices



#include <iostream>
using namespace std;

// --------------------------------------------------
// ESTRUCTURA: Nodo de lista enlazada
// --------------------------------------------------
struct Node {
    int data;   // Valor almacenado en el nodo

    Node* next; 
    // 🔥 PREGUNTA:
    // ¿Esto es un nodo?
    // ❌ No
    // 👉 Es una DIRECCIÓN a otro nodo
};


// --------------------------------------------------
// FUNCIÓN: Búsqueda en Linked List
// --------------------------------------------------
int buscar(Node* head, int target) {

    Node* current = head;   
    // 🔥 PREGUNTA:
    // ¿Estamos copiando la lista?
    // ❌ No
    // 👉 Solo copiamos la DIRECCIÓN (puntero)

    while (current != NULL) {
        // 🔥 PREGUNTA:
        // ¿Por qué NULL?
        // 👉 Porque indica fin de la estructura

        if (current->data == target) {
            // 🔥 PREGUNTA:
            // ¿Por qué -> ?
            // 👉 Porque current es puntero

            return 1;  // encontrado
        }

        current = current->next;
        // 🔥 PREGUNTA:
        // ¿Qué pasa si quitas esto?
        // 💀 loop infinito
    }

    return 0;  // no encontrado
}

// FUNCIÓN PRINCIPAL
// --------------------------------------------------

    Node* nuevo = new Node();

    nuevo_final->next = NULL;

    // 🔥 PREGUNTA:
